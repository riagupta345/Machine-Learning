### Normalization or Standardization:

- Normalized data has mean = 0 and standard deviation = 1 (or unit variance)
- If height is measured in km, std dev is very low as all values are very small => Should have SD = 1
- If height is measured in micrometers, std dev is very high as all values are very large => Should have SD = 1
- With u = 0 and SIG = 1, data is good for algorithm
- If training data is normalized, all incoming test data should also be normalized before applying to algorithm as input
- Normalizing data by:
 $\frac{x - \mu}{\sigma}$

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
dfx = pd.read_csv("../Datasets/Linear Regression/linearX.csv")
dfy = pd.read_csv("../Datasets/Linear Regression/linearY.csv")
```

```python
# Getting numpy arrays
x = dfx.values
y = dfy.values

# Normalizing data
x2 = (x - x.mean())/x.std()
```

```python
plt.scatter(x, y, label="Raw Data")
plt.scatter(x2, y, label="Normalized Data")    # => This has a mean of 0 (at 0 on graph's x-axis coordinate)
plt.legend()
plt.show()
```
