```python
# Create a map of movie title lengths vs frequency => How many movies have a title of given length

titles = df.get("movie_title")
# print(titles)
# print(type(titles))
```

```python
freq = {}

for t in titles:
    length = len(t)
    try:
        freq[length] += 1    # If freq value already exists
    except:
        freq[length] = 1     # If freq value doesnt exist, it throws exception, so comes here to create a new entry
        
print(freq)
```

```python
X = freq.keys()
Y = freq.values()
print(type(X), type(Y))
```

```python
X1 = np.array(list(X))
Y1 = np.array(list(Y))
# plt.bar(X1, Y1, width=5)

plt.scatter(X1, Y1)
plt.xlabel("Length of Movie Title")
plt.ylabel("Number of Movies")
plt.show()
```
