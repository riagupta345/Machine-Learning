### Scatters

```python
# To display only POINTS, not lines => Use scatter function
# Both x and y parameters are compulsory

plt.scatter(x, x)    # y = x line (as x-axis and y-axis hold same values of array x)
plt.scatter(x, y)    # y = 2x
plt.show()
```

```python
# Other parameters of scatter
plt.scatter(x, z, color = 'green')
plt.show()
```

```python
plt.scatter(x, z, marker = '^')
plt.show()
```

```python
plt.scatter(x, z, label = 'Stock Price')
plt.legend()                               # To display the box with label
plt.xlabel("Days")
plt.ylabel("Price")
plt.title("Stock Price Prediction")
plt.show()
```

```python
plt.plot(x, x**2, label = "2017")
plt.plot(x, x**3, label = "2018")
plt.legend()
plt.xlabel("Days")
plt.ylabel("Pricea")
plt.title("Stock Price Comparison")
plt.show()
```

### Bar Graph

```python
numbers = np.random.randint(0, 10, 5)
numbers2 = np.random.randint(0, 10, 5)
indices = np.arange(5)
print(numbers)
print(numbers2)
print(indices)
```

```python
plt.bar(indices, numbers)
plt.show()
```

```python
plt.bar(indices, numbers, 0.5, color="Orange", label="Gold")    # 0.5 is width
plt.legend()
plt.show()
```

```python
plt.bar(indices, numbers, 0.5, color="orange", label="Gold")
plt.bar(indices, numbers2, 0.5, color="green", label="Silver")

plt.legend()
plt.show()

# Doesnt look good, as green overlaps on orange
```

```python
plt.bar(indices, numbers, 0.5, color="orange", label="Gold")
plt.bar(indices+0.5, numbers2, 0.5, color="green", label="Silver")    
# Shift green bars by 0.5 by adding 0.5 to all array values

plt.legend()
plt.show()

# Still not pretty, so reduce width more
```

```python
plt.bar(indices, numbers, 0.25, color="orange", label="Gold")
plt.bar(indices+0.25, numbers2, 0.25, color="green", label="Silver")
plt.xlabel("Days")
plt.ylabel("Prices")
plt.title("Gold vs Silver Prices")

plt.legend()
plt.show()
```

### Pie Charts

```python
labels = ["Algebra", "Algorithms", "Python", "Calculus"]
values = [80, 90, 45, 70]    # Dont need to sum to 100 or 360, the reqd percentage is automatically calculated
```

```python
plt.pie(values, labels = labels, radius=1)
plt.show()
```

```python
# To highlight a wedge/section => Specify a tuple with 1 at the index of the section to be highlighted
# eg, to highlight Python, we give tuple as (0, 0, 1, 0)

plt.pie(values, labels=labels, radius = 1, explode=(0,0,1,0))
plt.show()
```

```python
# Bigger plot and with shadow

plt.pie(values, labels=labels, radius = 2, explode=(1,0,0,0), shadow=True)
plt.show()
```

### Histogram

#### 1D Normal Distribution

```python
# 1D Normal Distribution

# When u = 0 and sigma = 1, it is known as standard normal distribution
u = 5
sigma = 2

vals = u + sigma*np.random.randn(1000)
print(vals.shape)
# print(vals)
```

```python
plt.hist(vals, 50)    # 2nd parameter is bins => In how many categories the 1000 data points are added
plt.hist(vals, 100)
plt.show()

# plt.hist?
```

```python
plt.hist(vals, 10)
plt.show()

# When bins = 50, almost 60 data points are in middle most class
# When bins = 10, almost 250 data points are in middle most class
# Sum of y-axis values of each bar will sum upto 1000 in each case as we have 1000 data points
# On y-axis, P(x) = number of values falling in that interval x (probability density)
# Probability density is max around mean, and decreases as we move farther
```

```python
# To get count of values in each interval

vals = np.round(vals)
z = np.unique(vals, return_counts=True)
print(z)

# Sum is 1000, and 5 ke corresponding we have 204 (max value for mean and around it)
```

```python
a = np.random.randn(200)
plt.hist(a, 10)
plt.show()

# To show the data points generated:
x = a 
y = np.zeros(x.shape)     # To see values of x, y has to be zero
plt.scatter(x, y)
plt.show()
# Max data points around x = 0 (as mean is 0)
# 95% of the points lie between u-2sigma and u+2sigma
```

#### Multivariate Normal Distribution

```python
# 2D (Bivariate)

mean = np.array([0.0, 0.0])               # mean for our curve is at (0,0)
cov1 = np.array([[1, 0],[0, 1]]) # SIGxy = 0, ie, x and y are independent. Change in x causes no change in y

dist = np.random.multivariate_normal(mean, cov1, 500)
# print(dist)
print(dist.shape)    # (500, 2) as 500 points of 2 dimensions x and y => 2: first dim is x, second is y coordinate

plt.scatter(dist[:, 0], dist[:, 1])    
# On x-axis => all 500 points ka x => dist array ke all 500 points ka first component, similar for y
plt.show()
```

```python
mean2 = np.array([0.0, 0.0])               
cov2 = np.array([[1, 0.5], [0.5, 4]])     # More spread along y axis as SIGyy is more than SIGxx

dist2 = np.random.multivariate_normal(mean2, cov2, 500)
print(dist2.shape)   



mean3 = np.array([5.0, 6.0])                # Now centered at (5, 6) => max probability distribution
cov3 = np.array([[1.3, -0.9], [-0.9, 1.1]]) 

dist3 = np.random.multivariate_normal(mean3, cov3, 500)
print(dist3.shape)   

plt.scatter(dist2[:, 0], dist2[:, 1])    
plt.scatter(dist3[:, 0], dist3[:, 1])    
plt.show()

# Change cov array to see different manipulations of graph
# SIGxx increases => Spread along x-axis increases
# SIGyy increases => Spread along y-axis increases
# SIGxy increases => Dependence of x and y increases, so graph becomes oval in shape
```
