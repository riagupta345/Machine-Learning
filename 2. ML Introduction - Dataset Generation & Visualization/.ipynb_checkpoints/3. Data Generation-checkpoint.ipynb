## Generating Classification Data
Generating data for Binary Classification: Predict fruit based on size and sweetness 
- 2 features: size and sweetness
- 2 classes: apple and orange

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Generate two clusters of data, one cluster for data of apples and other cluster for data of oranges
# Specify mean and covariance of each cluster to define the cluster shape & size

u1 = np.array([2, 4])
cov1 = np.array([[2, -0.8], 
                 [-0.8, 1]])

u2 = np.array([-1, -3])
cov2 = np.array([[1, 0],
                 [0, 2]])
```

```python
x1 = np.random.multivariate_normal(u1, cov1, 300)    # Data of 300 apples
print(x1.shape)

x2 = np.random.multivariate_normal(u2, cov2, 400)    # Data of 400 oranges
print(x2.shape)
```

```python
plt.scatter(x1[:, 0], x1[:, 1], label="Apple")
plt.scatter(x2[:, 0], x2[:, 1], label="Orange")
plt.legend()
plt.xlabel('x1: Size')
plt.ylabel('x2: Sweetness')
plt.title("Apples and Oranges")
plt.show()
```

```python
# Now, we have two input arrays, one for apples and one for oranges
# For overall input array X, we must combine these two arrays together
# For output array Y, we must have a 700x1 array with value as class (say 1 for apple and 0 for orange)
```

```python
Y = np.zeros(700)
Y[ : 300] = 1     # From index 0 to 299 => mark 1 for apple, so 300 to 699 are 0 for orange
# print(Y)
print(Y.shape)
```

```python
X = np.vstack((x1, x2)) # Vertically stack the arrays: (n, ) arrays converted to (1, n) and stacked
print(X.shape)          # Overall input array => concatenation of both apple and orange arrays
```

```python
plt.scatter(X[:, 0], X[:, 1])
plt.show()
```

```python
# To change colour of each cluster => Assign cluster points individually

for i in range(700):
    if Y[i] == 1:
        plt.scatter(X[i, 0], X[i, 1], color='red')
    else:
        plt.scatter(X[i, 0], X[i, 1], color='orange')
        
plt.show()
```

```python
# Shortcut
plt.scatter(X[:, 0], X[:, 1], c=Y)    # c takes in sequence of values, where 0 and 1 indicate value of color
plt.show()
```
