## Generating datasets
Functions available in sklearn.datasets
    - make_regression()     => For linear regression
    - make_classification() => For classification
    - make_blobs()          => For clustering
    - make_moons()          => For non-linear data (neural networks, etc)

```python
from sklearn.datasets import make_regression, make_classification, make_blobs, make_moons
import matplotlib.pyplot as plt
```

### make_regression()

```python
X, Y = make_regression?
# X: Input array of features
# Y: Output array => Value is the regression value (of the equation as y = mx + c or sth)
```

```python
X, Y = make_regression(n_samples=200, n_features=1, n_informative=1)
print(X.shape, Y.shape)

# X is (200, 1) => 200 training samples of 1 feature
# Y is (200,)   => 200 entries wala vector containing its output value
```

```python
plt.scatter(X, Y)
plt.show()
```

```python
X1, Y1 = make_regression(n_samples=200, n_features=1, n_informative=1, noise = 8.8)
plt.scatter(X1, Y1)
plt.show()

# Gives different dataset each time
```

```python
X1, Y1 = make_regression(n_samples=200, n_features=1, n_informative=1, noise = 7, random_state=5)
plt.scatter(X1, Y1)
plt.show()

# Gives same dataset each time as random state is specified
```

### make_classification()

```python
XC, YC = make_classification(n_samples=200, n_features=2, n_informative=2, n_redundant=0)
print(XC.shape, YC.shape)
# print(YC)

# X => (200, 2) as 200 training examples with 2 features each
# Y => (200, ) as output vector has one entry per training input to denote the class => Stores class label as 
# 0/1 if number of classes=2
```

```python
plt.scatter(XC[:,0], XC[:, 1])
plt.show()
```

```python
# By default, number of classes is 2, number of clusters per class is 2

XC2, YC2 = make_classification(n_samples=200, n_features=2, n_informative=2, n_redundant=0, n_clusters_per_class=1)
plt.scatter(XC2[:,0], XC2[:,1], c=YC2) # To set colour sequence as class labels

# Only one cluster per class now
```

```python
XC2, YC2 = make_classification(n_samples=400, n_features=2, n_informative=2, n_redundant=0, n_clusters_per_class=1, random_state=5)
plt.scatter(XC2[:,0], XC2[:,1], c=YC2) # To set colour sequence as class labels
plt.show()

# Always generates same dataset as random state is specified (same dataset as CB video)
```

```python
XC3, YC3 = make_classification(n_samples=400, n_features=2, n_informative=2, n_redundant=0, n_classes=3, n_clusters_per_class=1, random_state=5)
plt.scatter(XC3[:,0], XC3[:,1], c=YC3) # To set colour sequence as class labels
plt.show()
```

### make_blobs()

```python
XB, YB = make_blobs?
```

```python
XB, YB = make_blobs(n_samples=200, n_features=2, centers=5)
plt.scatter(XB[:,0], XB[:,1], c=YB) 
plt.show()

# XB => Input samples
# YB => Integer labels for cluster membership of each sample
```

```python
XB, YB = make_blobs(n_samples=1000, n_features=2, centers=5, random_state=10)
plt.scatter(XB[:,0], XB[:,1], c=YB) 
plt.show()
```

### make_moons()
Non linear dataset

```python
XM, YM = make_moons(n_samples=100, noise=None, random_state=5)
plt.scatter(XM[:,0], XM[:,1], c=YM)
plt.show()
```

```python
# Generates two interleaving moons
# Cant separate dataset with linear line boundary
# Linear classifier cant be used
# Non linear classifiers can be used
```

```python
XM2, YM2 = make_moons(n_samples=400, noise=0.2, random_state=5)
plt.scatter(XM2[:,0], XM2[:,1], c=YM2)
plt.show()
```

### Subplots

```python
# 1: On same graph
plt.scatter(X1, Y1)
plt.scatter(XC2[:,0], XC2[:,1], c=YC2)
plt.show()
```

```python
# 2: Two separate figures
plt.figure()
plt.scatter(X1, Y1)
plt.figure(figsize=(8,8))
plt.scatter(XC3[:,0], XC3[:,1], c=YC3)
plt.show()
```

```python
# 3: In different sections/quadrants (subplots) of same figure
plt.figure(figsize=(8,8))

plt.subplot(221)
plt.title("Regression")
plt.scatter(X1, Y1)

plt.subplot(222)
plt.title("Classification")
plt.scatter(XC3[:,0], XC3[:,1], c=YC3)

plt.subplot(223)
plt.title("Blobs")
plt.scatter(XB[:,0], XB[:,1], c=YB)

plt.subplot(224)
plt.title("Moons")
plt.scatter(XM2[:,0], XM2[:,1], c=YM2)

plt.show()

# Once we call subplot function, everything after that belongs to that quadrant, thus we can have 4 separate titles
```
