# Multivariate Linear Regression

### 1. Univariate case:
$h_\theta (x) = \theta_0 + \theta_1x$ <br> 
where x is a single feature on which the output depends and $\theta_0$ is intercept


### 2. Multivariate case:
$h_\theta (x) = \theta_0 + \theta_1x_1 + \theta_2x_2 + \theta_3x_3$

- **Goal**: To learn best values of parameters $\theta_0$, $\theta_1$, $\theta_2$, $\theta_3 $
- For n features,  model has to learn n+1 parameters (including intercept $\theta_0$)

<br>**Generalization:** <br>
$h_\theta (x) = \theta_0 + \theta_1x_1 + \theta_2x_2$ $+$ $ ... $ $+$ $\theta_jx_j$  $+$ $ ... $ $+$ $\theta_nx_n$

- This represents a hyperplane in n-dimensions
    - 2 parameters: line
    - 3 parameters: plane in 3D
    - n parameters: hyperplane in nD
    

$h_\theta (x) = \theta_0x_0 + \theta_1x_1 + \theta_2x_2$ $+$ $ ... $ $+$ $\theta_jx_j$  $+$ $ ... $ $+$ $\theta_nx_n$, where $x_0$ = 1 always

$h_\theta (x) = \sum_{i=0}^{n} \theta_ix_i$, where $x_0$ = 1 always

<br>**Vector Representation:**
- Both vectors $\theta$ and $x$ are column vectors of dimensions $(n+1)\times 1$

Dataset X can be represented as shown in image <br>
- $x_j^i$ => i<sup>th</sup> example, j<sup>th</sup> feature <br><br>
- $x^{(i)}$ => All the features of i<sup>th</sup> example
