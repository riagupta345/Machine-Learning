```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
# a = [1, 2, 3]
# b = [4, 5, 6, 7]

a = np.arange(-4, 4, 0.02)    # Increasing number of points in a and b makes area of plotted squares smaller
b = a

a, b = np.meshgrid(a, b)

print(a)
print()
print(b)


# Prints 'a' row wise len(b) times     => Row [1 2 3] repeated 4 times       => Assigned to 'a'
# Prints 'b' column wise len(a) times  => Column [4 5 6 7] repeated 3 times  => Assigned to 'b'
```

```python
# Now, we choose same index elements from both arrays to get a list of coordinates:
# (4,1), (4,2), (4,3), (5,1), (5,2), (5,3), ...

m = a.shape[0]
n = a.shape[1]

print(m, n)

# for i in range(m):
#     for j in range(n):
#         print("(%f,%f)"%(b[i,j],a[i,j]), end=" ")
#     print()  
    
# Dont execute, takes too long
```

```python
# These mxn points will be computed by the 3D plotting function and plotted on a 3D surface, based on an elevation
# Elevation equation is provided by us as parameter

fig = plt.figure()
axes = fig.gca(projection='3d')
axes.plot_surface(a, b, a**2+b**2, cmap='rainbow')
plt.show()


# 3 parameters in plot_surface are: x, y, z coordinates
# So we plot (x, y, x2+y2) for a bowl shaped curve 
# (And also plotting squares of x and y as our error function depends on square of two parameters theta0 & theta1)
```

```python
# Contour plot
fig = plt.figure()
axes = fig.gca(projection='3d')
axes.contour(a, b, a**2+b**2, cmap='rainbow')
# plt.show()
```
