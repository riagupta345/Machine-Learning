### Limitations of K-Means
- K-Means fails when the data is not linearly separable, ie, non-linear data
- We use DBSCAN in such cases (Density-Based Spatial Clustering of Applications with Noise)

```python
from sklearn.datasets import make_moons
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
X, Y = make_moons(n_samples = 200, noise=0.1)
```

```python
plt.scatter(X[:,0], X[:,1], c=Y)
plt.show()
```

```python
km = KMeans(n_clusters=2)
km.fit(X, Y)
```

```python
centers = km.cluster_centers_
pred_labels = km.labels_
```

```python
plt.scatter(X[:,0], X[:,1], c=pred_labels)
plt.scatter(centers[:,0], centers[:,1], c='red', marker=',')
plt.show()
```

## DBSCAN
- Stands for Density-Based Spatial Clustering of Applications with Noise
- Finds core samples of high density and expands clusters from them
- Good for data which contains clusters of similar density.
- Idea is that a cluster is a region of high density surrounded by a region of low density
- It starts by exploring a small area, and if density of that region is good enough, it is considered as a part of the cluster. Then, it explores the neighbours, and continues grouping regions based on density
- SKLearn's algorithm does not take 'number of clusters' as a parameter, as clusters are made based on density only

```python
from sklearn.cluster import DBSCAN
```

```python
dbs = DBSCAN(eps=0.2, min_samples=5)
# eps: The MAXIMUM distance between two samples for them to be considered as in the same neighborhood
# Read docstring
```

```python
dbs.fit(X)
```

```python
# dbs.core_sample_indices_
```

```python
Y_pred = dbs.fit_predict(X)
```

```python
plt.scatter(X[:,0], X[:,1], c=Y_pred)
plt.show()
```
