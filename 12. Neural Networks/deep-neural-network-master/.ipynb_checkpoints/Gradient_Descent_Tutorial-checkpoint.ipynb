## What is Gradient Descent in Machine Learning? 
- Optimization Algorithm
- Based upon First Order Derivatives
- Most widely used technique starting from Linear Regression to Deep Learning
- Other optmization methods - Closed Form Solution, Second Order Derivates etc

 #### Why do we need this?
 - To reduce "COST/LOSS" while training any machine learning model
 - Loss Function is directly related to accuracy of the model
 - Zero Loss means 100% accuracy!

### Let Us see a very Simple Case!

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python

x = np.arange(10)

def lossFn(x):
    return (x-5)**2
    
loss = lossFn(x)
```

```python
plt.style.use("seaborn")
plt.plot(loss)
plt.show()
```

```python
def gradDescent():
    x_ = 0
    learning_rate = 0.1
    plt.plot(loss)
    
    for i in range(25):
        grad = 2*(x_-5)
        x_ = x_ - learning_rate*(grad)
        print("X_ %.4f "%x_)
        l = lossFn(x_)
        plt.scatter(x_,l)
    
```

```python
gradDescent()
```
