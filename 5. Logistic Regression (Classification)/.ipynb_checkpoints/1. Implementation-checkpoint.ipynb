### Data Preparation (Generation)

```python
mean1 = np.array([1, 0.5])
cov1 = np.array([[1, 0.1], [0.1, 1.2]])

mean2 = np.array([4, 5])
cov2 = np.array([[1.21, 0.1], [0.1, 1.3]])
```

```python
# 2 datasets with normal distribution and 2 features each

dist1 = np.random.multivariate_normal(mean1, cov1, 500)
dist2 = np.random.multivariate_normal(mean2, cov2, 500)

print(dist1.shape, dist2.shape)
```

```python
X = np.vstack((dist1, dist2))
print(X.shape)

Y = np.zeros(X.shape[0])
Y[500:] = 1
print(Y.shape)
Y = Y.reshape((1000,1))

data = np.hstack((X,Y))
print(data.shape)

np.random.shuffle(data)
# print(data2[:501])
```

```python
split = int(0.8*data.shape[0])
```

```python
X_train = data[:split, :-1]
X_test = data[split:, :-1]

Y_train = data[:split, -1]
Y_test = data[split:, -1]

print(X_train.shape,X_test.shape)
print(Y_train.shape,Y_test.shape)
```

## Logistic Regression Functions

```python
def hypothesis(x, w, b):             # Input vector x, theta (weight) vector w, bias 
    h = np.dot(x,w) + b
    return sigmoid(h)

def sigmoid(z):
    return 1.0/(1.0 + np.exp(-1.0*z))

def error(X, Y_true, w, b):
    m = X.shape[0]
    err = 0.0
    for i in range(m):
        hx = hypothesis(X[i], w, b)
        err += Y_true[i]*np.log2(hx) + (1 - Y_true[i])*np.log2(1-hx)
    return -err/m      # Return average error

def get_grad(X, Y_true, w, b):
    grad_w = np.zeros(w.shape)
    grad_b = 0.0
    m = X.shape[0]
    for i in range(m):
        hx = hypothesis(X[i], w, b)
        grad_w += (Y_true[i] - hx)*X[i]
        grad_b += (Y_true[i] - hx)
    grad_w /= m
    grad_b /= m
    return [grad_w, grad_b]

def gradientDescent(X, Y_true, w, b, lr = 0.1):       # One iteration of gradient descent
    err = error(X, Y_true, w, b)
    [grad_w, grad_b] = get_grad(X, Y_true, w, b)
    w = w + lr*grad_w
    b = b + lr*grad_b
    return err, w, b
    
def predict(x, w, b):
    confidence = hypothesis(x, w, b)
    if confidence < 0.5:
        return 0
    else:
        return 1
    
def get_accuracy(x_test, y_test, w, b):
    y_predicted = []
    for i in range(y_test.shape[0]):
        p = predict(x_test[i], w, b)                          # Predict on x_test
        y_predicted.append(p)
        
    y_predicted = np.array(y_predicted)
    return float((y_predicted == y_test).sum())/y_test.shape[0]    # Compare prediction with actual y_test
```

```python
W = 2*np.random.random((X_train.shape[1],))
B = 5*np.random.random()

print(W)
print(B)

loss = []
accuracy = []
```

```python
for i in range(5000): # 1000 epochs
    ls, W, B = gradientDescent(X_train, Y_train, W, B, lr=0.1)
    loss.append(ls)
    accuracy.append(get_accuracy(X_test, Y_test, W, B))
```

```python
plt.plot(loss)
plt.ylabel("Negative of Log Likelihood")
plt.xlabel("Time")
# plt.xlim(-50, 800)
plt.show()
```

```python
plt.plot(accuracy)
plt.show()
print(accuracy[-1])
```
