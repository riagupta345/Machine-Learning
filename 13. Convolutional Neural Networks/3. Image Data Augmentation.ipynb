#### Need for Data Augmentation:
- Deep learning models are "data hungry" => need lots of data to learn features
- Often lead to overfitting when datasets are small
- So, we can generate more number of training samples by augmenting the training samples
- Makes model more robust and leads to increased accuracy

```python
img_path = "../13. Convolutional Neural Networks/ria_nz.jpg"
```

```python
import matplotlib.pyplot as plt
import cv2
```

```python
img = cv2.imread(img_path)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
# plt.figure(figsize=(12,8))
plt.imshow(img)
```

```python
img.shape
```

```python
from keras.preprocessing import image
```

```python
img2 = image.load_img(img_path)
```

```python
x = image.img_to_array(img2)/255.0
```

```python
x.shape
```

```python
plt.imshow(x)
```

```python
x = x.reshape((1, 2048,1662,3))
```

```python
data_gen = image.ImageDataGenerator(rotation_range=30, height_shift_range=0.2, width_shift_range=0.2, shear_range=0.2,
                                    zoom_range=0.2, horizontal_flip=True, fill_mode='nearest')
```

```python
i = 0

for batch in data_gen.flow(x, batch_size=1):
    plt.figure()
    plt.imshow(image.img_to_array(batch[0]))
    i += 1
    plt.axis("off")
    plt.show()
    if i == 5:
        break
print(i)
```
